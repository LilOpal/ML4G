```python
data_prefix = './datasets_knowledge_embedding/'
dataset_names = ['FB15K/', 'FB15k-237/', 'WN18/text/', 'WN18RR/text/']
dataset_division_names = ['train.txt', 'valid.txt', 'test.txt']

for dataset_name in dataset_names:
    output_name = dataset_name.split('/')[0]
    print(output_name)
    entity_set = set()
    rel_set = set()
    for div_name in dataset_division_names:
        dataset_path = data_prefix + dataset_name + div_name
        print(div_name)
        with open(dataset_path, 'r') as f:
            lines = f.readlines()
            
        h_set = set()
        r_set = set()
        t_set = set()
        
        print(f'Num of Triples: {len(lines)}')
        for line in lines:
            h, r, t = line.split()
            h_set.add(h)
            r_set.add(r)
            t_set.add(t)
        rel_set.update(r_set)
        entity_set = entity_set.union(t_set)
        entity_set = entity_set.union(h_set)
        
        print(f'Num of Relation: {len(r_set)}')
        print(f'Num of Head Entities: {len(h_set)}')
        print(f'Num of Tail Entities: {len(t_set)}')
        print('')
    print(f'Num of all Entity: {len(entity_set)}')
    print(f'Num of all Relation: {len(rel_set)}')
    print('\n')
```
